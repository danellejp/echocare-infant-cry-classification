```python
import tensorflow as tf
import numpy as np
import os
from datetime import datetime

print(f"TensorFlow version: {tf.__version__}")
```

```python
# Load classification model
classification_model_path = 'original_models/best_model_classification.keras'

print("Loading classification model...")
classification_model = tf.keras.models.load_model(classification_model_path)

print("Model loaded successfully")
print(f"Model input shape: {classification_model.input_shape}")
print(f"Model output shape: {classification_model.output_shape}")
```

```python
# Convert to TFLite
print("\nConverting to TFLite format...")

converter = tf.lite.TFLiteConverter.from_keras_model(classification_model)
tflite_model = converter.convert()

tflite_path = 'tflite_models/classification_model.tflite'
with open(tflite_path, 'wb') as f:
    f.write(tflite_model)

file_size = os.path.getsize(tflite_path) / (1024 * 1024)
print(f"Conversion complete")
print(f"Saved to: {tflite_path}")
print(f"File size: {file_size:.2f} MB")
```
