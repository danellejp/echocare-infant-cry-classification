```python
import tensorflow as tf
import numpy as np
import os
from datetime import datetime

print(f"TensorFlow version: {tf.__version__}")
print(f"Conversion started at: {datetime.now()}")
```

```python
# Load Detection Model
detection_model_path = 'original_models/best_model_detection.keras'

print("Loading detection model...")
detection_model = tf.keras.models.load_model(detection_model_path)

print("Model loaded successfully")
print(f"Model input shape: {detection_model.input_shape}")
print(f"Model output shape: {detection_model.output_shape}")
```

```python
# Convert to TFLite
print("\nConverting to TFLite format...")

# Create converter
converter = tf.lite.TFLiteConverter.from_keras_model(detection_model)

# Convert
tflite_model = converter.convert()

# Save
tflite_path = 'tflite_models/detection_model.tflite'
with open(tflite_path, 'wb') as f:
    f.write(tflite_model)

# Check file size
file_size = os.path.getsize(tflite_path) / (1024 * 1024)  # Convert to MB
print(f"Conversion complete")
print(f"Saved to: {tflite_path}")
print(f"File size: {file_size:.2f} MB")
```
